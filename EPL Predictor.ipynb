```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt 
%matplotlib inline
```

```python
epl=pd.read_csv("epl2020.csv")
```

```python
teams=epl["teamId"].unique()
table=[]
for i in teams:
    team_data=epl[epl["teamId"]==i]
    Won=team_data["wins"].sum()
    Drawn=team_data["draws"].sum()
    Lost=team_data["loses"].sum()
    Played=Won+Drawn+Lost
    GF=team_data["scored"].sum()
    GA=team_data["missed"].sum()
    GD=GF-GA
    Points=team_data["pts"].sum()
    table.append([i,Won,Drawn,Lost,Played,GF,GA,GD,Points])
league=pd.DataFrame(table,columns=["Club","Won","Drawn","Lost","Played","GF","GA","GD","Points"])
league.sort_values(by="Points",ascending=False,inplace=True)
league.set_index(np.arange(1,21),inplace=True)
league
```

```python
#Dirty Clubs (2019/2020)
teams=epl["teamId"].unique()
table=[]
for i in teams:
    cards=epl[epl["teamId"]==i]
    Hred=cards["HR.x"].sum()
    Ared=cards["AR.x"].sum()
    red=Hred+Ared
    Hyel=cards["HY.x"].sum()
    Ayel=cards["AY.x"].sum()
    yel=Hyel+Ayel
    Hfoul=cards["HF.x"].sum()
    Afoul=cards["AF.x"].sum()
    fouls=Hfoul+Afoul
    table.append([i,red,yel,fouls])
```

```python
dirty=pd.DataFrame(table,columns=["Club","RedCards","YellowCards","Fouls"])
dirty.sort_values(by=["RedCards","YellowCards","Fouls"],ascending=False,inplace=True)
dirty.set_index(np.arange(1,21),inplace=True)
dirty.head(20)
# Aston Villa has committed the most no of fouls so for this season (2019/2020)
```

```python
epl["GF"]=epl["xG"].apply(lambda x:round(x))
epl["GA"]=epl["xGA"].apply(lambda x:round(x))
```

```python
epl['xwins'] = epl.apply(lambda x: 1 if x['GF'] > x['GA'] else 0, axis=1)
epl['xlosses'] = epl.apply(lambda x: 1 if x['GF'] < x['GA'] else 0, axis=1)
epl['xdraws'] = epl.apply(lambda x: 1 if x['GF'] ==x['GA'] else 0, axis=1)
```

```python
epl["xlosses"].sum()
```

```python
teams=epl["teamId"].unique()
xtable=[]
for i in teams:
    xepl=epl[epl["teamId"]==i]
    
    wins=xepl["xwins"].sum()
    losses=xepl["xlosses"].sum()
    draws=xepl["xdraws"].sum()
    points=(3*wins)+(1*draws)
    xtable.append([i,wins,losses,draws,points])

xleague=pd.DataFrame(xtable,columns=["Club","Wins","Losses","Draws","Points"])
xleague.sort_values(by="Points",ascending=False, inplace=True)
xleague.set_index(np.arange(1,21),inplace=True)
xleague

# Considering the expected metrics, ManCity is expected to score the highest points followed by Liverpool FootBall Club
```

```python
# Recent Form
teams=epl["teamId"].unique()
form=[]
for i in teams:
    recent=epl[epl["teamId"]==i].tail(10)
    rwins=recent["wins"].sum()
    rdraws=recent["draws"].sum()
    points=(3*rwins)+(1*rdraws)
    form.append([i,points])
recentform=pd.DataFrame(form, columns=["Club","Points"])
recentform.sort_values(by="Points", ascending=False,inplace=True)
recentform.set_index(np.arange(1,21))
```

```python
teams=epl["teamId"].unique()
Points=[]
for i in teams:
    clubs=epl[epl["teamId"]==i]
    count=len(clubs.index)
    xPoints=clubs["pts"].sum()
    PPP=xPoints/count
    Remaining_games=38-count
    Predicted_points=int(round(xPoints+Remaining_games*PPP))
    Points.append([i,Predicted_points])
Final_Standings=pd.DataFrame(Points,columns=["Clubs","Estimated Points"])
Final_Standings.sort_values(by="Estimated Points", ascending=False,inplace=True)
Final_Standings.set_index(np.arange(1,21))
# Considering the points per game model, Liverpool should win the 2019/2020 Title(Champions)
```
